```python
import pandas as pd

# Lista de diccionarios con los datos
data = []

# Recorremos cada ronda y extraemos los datos
for round_idx, round_data in enumerate(rounds):
    round_list = []
    for player, stats in round_data.items():
        round_list.append({
            'Player': player,
            'Kills': stats['kills'],
            'Assists': stats['assists'],
            'Deaths': stats['deaths']
        })
    
    # Convertimos la lista en un DataFrame y mostramos por ronda sin índice
    df_round = pd.DataFrame(round_list)
    print(f"Round {round_idx + 1}")
    print(df_round.to_string(index=False))
    print("\n")
```
